```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from pathlib import Path

DATA_RAW = Path('../data/raw')
DATA_PROCESSED = Path('../data/processed')
FIGURES = Path('../figures')

plt.rcParams['figure.figsize'] = (12, 5)
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.right'] = False
```

## 1. Load Data

```python
# Load your dataset here
# df = pd.read_csv(DATA_RAW / 'companies.csv')
# df.head()

# Placeholder — replace with real data
df = pd.DataFrame({
    'company': ['Company A', 'Company B', 'Company C', 'Company D', 'Company E'],
    'sector': ['Synthetic Biology', 'Fermentation', 'AI Drug Discovery', 'Synthetic Biology', 'Fermentation'],
    'funding_m': [12, 45, 120, 8, 67],
    'founded': [2019, 2017, 2021, 2020, 2018],
    'country': ['Netherlands', 'Germany', 'USA', 'UK', 'Netherlands'],
})
df
```

## 2. Sector Overview

```python
sector_funding = df.groupby('sector')['funding_m'].sum().sort_values(ascending=False)

fig = px.bar(
    sector_funding.reset_index(),
    x='sector', y='funding_m',
    title='Total Funding by Sector (€M)',
    color='sector',
    color_discrete_sequence=px.colors.qualitative.Set2,
)
fig.update_layout(showlegend=False, plot_bgcolor='white')
fig.write_image(str(FIGURES / 'sector_funding.png'))
fig.show()
```

## 3. Key Insights

```python
print(f"Companies tracked:   {len(df)}")
print(f"Total funding:       €{df['funding_m'].sum()}M")
print(f"Median funding:      €{df['funding_m'].median()}M")
print(f"Sectors covered:     {df['sector'].nunique()}")
print(f"Countries:           {df['country'].nunique()}")
```
